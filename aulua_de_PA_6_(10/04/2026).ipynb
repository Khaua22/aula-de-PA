```python
# escrever estrutura de decisões (if,else)
# escrevere programa que  diga se o numero é impar ou par

numero = int(input("digite um numero"))

#processamento/saída

if numero % 2 == 0:
    print(f"o {numero} é par")

else:
    print(f"o {numero} é ímpar")
```

```python
# Escrever um programa que leia o valor da compra e aplique:

compra = int(input("digite o valor da compra" ))

# Processamento / saída

if compra>=500:
  desconto = compra - (compra * 0.10)
  print(f"o valor a pagar é: {desconto}" )
elif compra >=101:
  desconto = compra - (compra * 0.20)
  print(f"o valor a pagar é: {desconto}")
else:
  print(f"sem desconto: {compra}")
```
