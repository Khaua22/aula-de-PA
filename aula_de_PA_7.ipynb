```python
#contar letras, o usuario digita uma palavra e o programa conta as letras da palavra

palavra = input("Digite uma palavra: ")
numero_letras = len(palavra)
print(f"A palavra '{palavra}' tem {numero_letras} letras.")
```

```python
#contar vogais, o usuario digita uma palavra e o programa conta as vogais
palavra = input("Digite uma palavra: ").lower()
vogais = "aeiou"
contador_vogais = 0
for letra in palavra:
  if letra in vogais:
    contador_vogais += 1
print(f"A palavra '{palavra}' tem {contador_vogais} vogais.")
```

```python
#maior numero digitado, o usuario digita quatro numeros e o programa imprimi o maior numero numero digitado

num1 = int(input("Digite o primeiro número: "))
num2 = int(input("Digite o segundo número: "))
num3 = int(input("Digite o terceiro número: "))
num4 = int(input("Digite o quarto número: "))

maior_numero = max(num1, num2, num3, num4)

print(f"O maior número digitado é: {maior_numero}")
```

```python
#o usuario digita um numero entre 01 e 100 e o programa mostra quais são divisiveis por 3 e o programa mostra quais são divisiveis por 3 e 5

while True:
    try:
        numero = int(input("Digite um número entre 1 e 100: "))
        if 1 <= numero <= 100:
            break
        else:
            print("Por favor, digite um número dentro do intervalo de 1 a 100.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

print(f"\nNúmeros divisíveis por 3 até {numero}:")
for i in range(1, numero + 1):
  if i % 3 == 0:
    print(i)

print(f"\nNúmeros divisíveis por 3 e 5 até {numero}:")
for i in range(1, numero + 1):
  if i % 3 == 0 and i % 5 == 0:
    print(i)
```
