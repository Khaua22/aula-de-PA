```
#Senha correta

#print("Cadastre aqui sua senha com os seguintes criétios: \n"
#      "         *Ao menos 8 digitos\n"
#      "         *Ao menos uma letra MAIÚSCULA\n"
#      "         *Ao menos um número\n"
#      "         *Ao menos um caractere especial(!@#$%¨&*)\n")

#validação

while True:
    senha = str(input("Digite sua senha : "))
    if senha.islower():
        print("A senha deve ter pelo menos um caractere MAIUSCULO: ")
    elif len(senha) < 7 :
        print("A senha deve ter pelo menos 8 caracteres: ")
    elif senha.isalpha() :
        print("Necessita de um numero: ")
    elif senha.isalnum() :
        print("Necessita de um Caractere especial: ")
    else:
        break
```

```
# Calculadora simples

# Entrada

print ("==========CALCULADORA==========|")
print ("==========MENU==========|")
print("                               |")
print("opção = 1 - Soma               |")
print("opção = 2 - Subtração          |")
print("opção = 3 - Multiplicação      |")
print("opção = 4 - Divisão            |")

# Entrada
numero = float (input("Digite um numero: "))
opcao = int (input("Escolha a opção desejada: "))
numero1 = float(input("digite um numero: "))

# Processamento / saida

if opcao == 1:
  print (f"A soma de {numero} e {numero1} é igual a {numero + numero1}")

elif opcao == 2:
  print (f"A subtração de {numero} e {numero1} é igual a {numero - numero1}")

elif opcao == 3 :
   print (f"A multiplicação de {numero} e {numero1} é igual a {numero * numero1}")

elif opcao == 4 :
   print (f"A divisão de {numero} e {numero1} é igual a {numero / numero1}")
```

```
# Dia da semana

# Leitura do número inteiro de 1 a 7
dia = int(input("Digite um número de 1 a 7: "))

# Verificação do número e exibição do dia correspondente
if dia == 1:
    print("Domingo")
elif dia == 2:
    print("Segunda-feira")
elif dia == 3:
    print("Terça-feira")
elif dia == 4:
    print("Quarta-feira")
elif dia == 5:
    print("Quinta-feira")
elif dia == 6:
    print("Sexta-feira")
elif dia == 7:
    print("Sábado")
else:
    print("Número inválido")
```

```
# Verifica se um ano é bissexto

# Entrada
ano = int(input("Ano? "))

# Processamento / Saída
if (ano % 400 == 0) or ((ano % 4 == 0) and (ano % 100 != 0)):
    print(f"{ano} é bissexto!")
else:
    print(f"{ano} não é bissexto!")
```
